## 作業目標：搭建一個bag of words模型

---

```python
import pandas as pd
import nltk
#nltk.download()
import numpy as np
dataset=pd.read_csv('Restaurant_Reviews.tsv',delimiter='\t',quoting=3)#  delimiter代表定位點
corpus=dataset['Review'].values
```

### 從文本中取出所有單字

```python
import nltk
nltk.download('punkt')
```

```python
whole_words = []
for sentence in corpus:
    tokenized_sentence =nltk.word_tokenize(sentence)#進行拆解
    for word in tokenized_sentence:
        whole_words.append(word)
```

### 移除重複單字

```python
whole_words=set(whole_words)#移除重複字詞
print('共有{}個單字'.format(len(whole_words)))#計算共有幾個字詞
```

### 建立字典使每一個單字有對應數值

```python
word_index = {}
index_word = {}
n = 0
for word in whole_words:
#     """自行填入"""
    word_index[word]=n
    index_word[n]=word
    n+=1                         #n=n+1

    
```
